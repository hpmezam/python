```python
import math
def calcularDistancia(x2, x1, y2, y1):
    dx = x2 - x1
    dy = y2 - x1
    cuadrado = dx**2 + dy**2
    distancia = math.sqrt(cuadrado)
    return distancia

x2, x1, y2, y1 = 1, 2, 4, 6
distancia = calcularDistancia(x2,x1,y2,y1)
print(f'distancia es igual a: {distancia}')

```
